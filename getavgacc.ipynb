```python
import json
import math
# filename=''
# filename='/home/chenruiyang/Code/LLM/EAGLE3/mt_bench_prune/llama38b2_40-temperature-0.0-pct025.jsonl'
filename='/home/chenruiyang/Code/LLM/EAGLE3/mt_bench_prune/llama38b2_40-temperature-0.0-pct04.jsonl'


def getacc(filename):
    # 从文件中读取数据
    with open(filename, 'r') as f:
        data = [json.loads(line) for line in f]

    # 提取所有ppl值
    all_ppl = []
    for entry in data:
        for choice in entry['choices']:
            all_ppl.extend(choice['acc'])

    # 计算算术平均值
    arithmetic_avg = sum(all_ppl) / len(all_ppl)

    # 输出结果
    # print("所有ppl值:", all_ppl)
    print(f"平均acc: {round(arithmetic_avg, 4)}")
    # print(f"")
    # print(f"几何平均比算术平均低: {round(arithmetic_avg - geometric_avg, 4)}")

# 示例调用
# getppl('your_file.json')
```

```python
import json

import re
ANS_RE = re.compile(r"#### (\-?[0-9\.\,]+)")
INVALID_ANS = "[invalid]"

def extract_final_answer(text):
    numbers = re.findall(r'\d+', text)
    return numbers[-1] if numbers else None

def extract_answer_gsm8k(completion):
    match = ANS_RE.search(completion)
    if match:
        match_str = match.group(1).strip()
        match_str = match_str.replace(",", "")
        return match_str
    else:
        return INVALID_ANS


with open('/home/chenruiyang/Code/LLM/EAGLE3/eagle/data/gsm8k/question.jsonl', 'r') as f:
    stdans=[]
    for i,line in enumerate(f):
        stdans.append(extract_answer_gsm8k(json.loads(line)['reference'][0]))

def getacc2(filename):
    # 从文件中读取数据
    with open(filename, 'r') as f:
        data = [json.loads(line) for line in f]

    # 提取所有ppl值
    all_ppl = []
    for entry in data:
        for choice in entry['choices']:
            all_ppl.extend(choice['turns'])

    accsum = 0
    for txt,stda in zip(all_ppl,stdans):
        # print(txt)
        # print('STDA:',stda)
        if not stda in txt:
            print(stda)
        accsum += stda in txt
    

    # 输出结果
    # print("所有ppl值:", all_ppl)
    print(f"平均acc: {round(accsum/len(all_ppl), 4)}")

# print(data)
```
